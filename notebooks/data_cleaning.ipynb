## Data cleaning:
### Converting `ApprovedBudgetForContract` and `ContractCost` to float:

```python
# Create a copy of the original dataframe
df_clean = df.copy()
```

```python
# Convert ApprovedBudgetForContract to float
df_clean['ApprovedBudgetForContract'] = pd.to_numeric(
    df_clean['ApprovedBudgetForContract'],
    errors='coerce'
)

# Convert ContractCost to float
df_clean['ContractCost'] = pd.to_numeric(
    df_clean['ContractCost'],
    errors='coerce'
)
```

```python
# Confirm successful conversion
df_clean[['ApprovedBudgetForContract', 'ContractCost']].dtypes
```

```python
# Count missing values after conversion
myca_budget_count = df_clean['ApprovedBudgetForContract'].isna().sum()
myca_cost_count = df_clean['ContractCost'].isna().sum()

print(f"ApprovedBudgetForContract missing after conversion: {myca_budget_count}")
print(f"ContractCost missing after conversion: {myca_cost_count}")
```

```python
# Percentage of affected records
(
    df_clean[['ApprovedBudgetForContract', 'ContractCost']]
    .isna()
    .mean()
    .mul(100)
    .round(2)
)
```

During data cleaning, the `ApprovedBudgetForContract` and `ContractCost` fields were converted from object to numeric data types. A review of conversion errors revealed that some records contained MYCA (Multi-Year Contract Authority) references instead of monetary values. These entries were converted to NaN using pd.to_numeric(errors='coerce'). Approximately 157 budget records and 28 contract cost records were affected and retained as missing values for subsequent analysis. All 28 invalid contract cost records also had invalid approved budget values, indicating a consistent data-entry pattern rather than isolated errors.

### Converting `ActualCompletionDate` and `StartDate` to datetime format:

```python
# Convert ActualCompletionDate and StartDate to datetime format
date_columns = ["ActualCompletionDate", "StartDate"]

for col in date_columns:
    df_clean[col] = pd.to_datetime(df_clean[col], errors="coerce")

# Verify the conversion
df_clean[date_columns].dtypes
```

Converted ActualCompletionDate and StartDate to datetime format to enable date-based analysis, filtering, and project timeline calculations.
